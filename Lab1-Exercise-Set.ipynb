```python
def validate(seq):
    isdna = True
    isrna = True
    seq = seq.upper()
    for i in seq:
        if i not in ['A', 'C', 'G', 'T']: isdna = False
        if i not in ['A', 'C', 'G', 'U']: isrna = False
    return "dna" if isdna else "rna" if isrna else "invalid"

print(validate("ACcAccaatCCGgTAA"))
print(validate("gguUgCuCGaaAuccaauu"))
print(validate("ACcCHHHHHHHCGCTTAA"))
```

```python
def rnadna(seq):
    return "dna" if "T" in seq.upper() else "rna"


print(rnadna("ggCGGcagacgagtcGCAgTccta"))
print(rnadna("ggGUaCUGUACugaCugacguucCgUAAa"))
```

```python
def occurs1(seq):
    seq = seq.lower()
    return [
        ['a', seq.count('a')],
        ['c', seq.count('c')],
        ['g', seq.count('g')],
        ['t', seq.count('t')]
    ]


print(occurs1("agCggtTGAct"))
```

```python
def occurs2(seq):
    seq = seq.lower()
    l = [['a', 0], ['c', 0], ['g', 0], ['t', 0]]
    for i in seq:
        if i == 'a': l[0][1] += 1
        elif i == 'c': l[1][1] += 1
        elif i == 'g': l[2][1] += 1
        elif i == 't': l[3][1] += 1
    return l


print(occurs1("agCggtTGAct"))
```

```python
def cgratio(seq):
    seq = seq.lower()
    return \
        (seq.count('c') + seq.count('g'))\
        / len(seq)


print(cgratio("cggcggGttaaattg"))
```

```python
def subseq(seq, pat):
    seq = seq.lower()
    pat = pat.lower()
    occurs = []
    for i in range(len(seq) - len(pat) + 1):
        if seq[i:i + len(pat)] == pat:
            occurs.append(i)
    return occurs, len(occurs)


print(subseq("actcgtgtccgtaacgt", "cgt"))
```

```python
def subseqratio(seq, pat):
    return (subseq(seq, pat)[1] * len(pat)) / len(seq)


print(subseqratio("gcgattgc", "g"))
```

```python
def kmers(seq, length):
    kmers_dict = {seq[0:length]: 1}
    for i in range(1, len(seq) - length + 1):
        kmer = seq[i:i + length]
        if kmer not in kmers_dict: kmers_dict[kmer] = 1
        else: kmers_dict[kmer] += 1
    return kmers_dict


for kmer in kmers("ACTGATCTATAAACTATCATCACTCTATACT", 3).items():
    print(kmer)
```

```python
def kmers_w_indices(seq, length):
    kmers_dict = {seq[0:length]: [1, [0]]}
    for i in range(1, len(seq) - length + 1):
        kmer = seq[i:i + length]
        if kmer not in kmers_dict: kmers_dict[kmer] = [1, [i]]
        else: kmers_dict[kmer][0] += 1; kmers_dict[kmer][1].append(i)
    return kmers_dict


for kmer in kmers_w_indices("ACTGATCTATAAACTATCATCACTCTATACT", 3).items():
    print(kmer)
```
