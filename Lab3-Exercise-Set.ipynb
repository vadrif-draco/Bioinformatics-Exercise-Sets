```python
def reconstruct_from_kmers(kmers_list):
    seq = ''
    for kmer in kmers_list[:-1]: seq += kmer[:1]
    seq += kmers_list[-1]
    return seq


print(reconstruct_from_kmers(['ATC' , 'TCG' , 'CGA' , 'GAC' , 'ACC']))
```

```python
#@title Default title text { display-mode: "code" }
variable_name = ""
def __lcs(seq1, seq2):

    # Generate grid of zeroes with dimensions of (len(seq1) + 1)Ã—(len(seq2) + 1)
    grid = [[0 for col in range(len(seq2) + 1)] for row in range(len(seq1) + 1)]

    # Variables to keep track of LCS so far, its length, heuristics, and logs
    lcs = ''; lcs_length_so_far = 0; curr_h = 0; prev_h = 0; logs = []

    # Loop over entire grid with indices i, j
    # Starting from the 2nd row and column
    for i in range(1, len(seq1) + 1):
        for j in range(1, len(seq2) + 1):

            # If the letters are matching
            if (seq1[i-1] == seq2[j-1]):

                # Add one to corresponding cell grid
                # This resembles incrementing the LCS with this new match
                grid[i][j] = grid[i-1][j-1] + 1
                
                # Set the current value for the heuristic used in ordering LCS
                # For this heuristic, lower is better
                curr_h = i+j+abs(i-j);

                # Check if no matches already happened in this position before
                # Where if so increment the LCS length and add the match to it
                if (grid[i-1][j-1] == lcs_length_so_far):
                    prev_h = curr_h
                    lcs += seq1[i-1]
                    lcs_length_so_far += 1
                    logs.append(f'i={i} j={j} h={curr_h} {lcs}')

                # Otherwise; if a match has already happened in this position
                # Check if it is more suitable than the one chosen earlier
                # If it is, replace it
                elif (curr_h < prev_h):
                    prev_h = curr_h
                    lcs = lcs[:-1] + seq1[i-1]
                    logs.append(f'i={i} j={j} h={curr_h} {lcs}')

            # Otherwise; if the letters are not matching
            else:

                # Get the max between left and top
                # This resembles proceeding with the current LCS so far
                grid[i][j] = max(grid[i-1][j], grid[i][j-1])

    return lcs, lcs_length_so_far, grid, logs


def lcs(seq1, seq2, debugging=False):
    lcs, lcslen, grid, logs = __lcs(seq1, seq2)
    if not debugging: return lcs
    else:
        out =(f'\n\n\nLCS: {lcs}, of length: {lcslen}\n')
        out+=("\nLogs:\n")
        for log in logs: out+=(f'{log}\n')
        out+=("\nGrid:\n")
        grid[0].insert(0, ' ')
        for i, row in enumerate(grid[1:]): row.insert(0, seq1[i])
        grid.insert(0, [' ', ' '] + [i for i in seq2])
        for row in grid:
            for item in row:
                out+=(f' {item} ')
            out+=('\n')
        out+=("\n\n\n")
        return out


print(lcs("GXTXAYB", "AGGTAB")) # GTAB
print(lcs("AGGTAB", "GXTXAYB")) # GTAB
print(lcs("ABCDEF", "CXABEF", debugging=True)) # ABEF
print(lcs("CXABEF", "ABCDEF", debugging=True)) # ABEF
print(lcs("AXCXTXG", "ACGG")) # ACG
print(lcs("ABXXEF", "ABFCD")) # ABF
print(lcs("ABFCD", "ABXXEF")) # ABF
print(lcs("ABCD", "ABCD")) # ABCD
print(lcs("AXXBCD", "ABCD")) # ABCD
print(lcs("ABCD", "AXXBCD")) # ABCD
print(lcs("ABYYYCD", "AXXBCDYX")) # ABCD
print(lcs("AXXBCDYX", "ABYYYCD")) # ABCD


# Problem in test case 2
# Trace prints show that the issue is there is no sense of which match came 1st
# That leads to 'A' (1, 5) matching first before the 'G' (2, 1)
# ------------------------------------------------------------------------------
# My proposed fix is to use the heuristic as a way to override our choice
# Where for the same position in the sequence, lower heuristic is preferred
# So, if a new match appears in the same position, I don't discard it directly
# Instead, I evaluate its heuristic value and see if it is better


# Problem in test cases 3 and 4
# My initial heuristic solved test case 2's problem
# However, the chosen heuristic breaks some cases with inverted priority
# For example, without the heuristic cases 3 and 4 return ABEF correctly
# But with it, they now favor 'C' (1, 3) over 'B' (2, 4) in different positions
# ------------------------------------------------------------------------------
```
