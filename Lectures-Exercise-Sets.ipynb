```python
def checkseqfreq(seq):
    seq = seq.upper()
    return {
        'A':seq.count('A'),
        'T':seq.count('T'),
        'C':seq.count('C'),
        'G':seq.count('G')
    }
    
    
print(checkseqfreq(seq))
```

```python
seq1 = 'GATTACA'
seq2 = 'GACTATA'

def checkseqhamming(seq1, seq2): # Assuming same length (which is a condition for hamming distance method which allows only for substitution type of differences)
    hammingdistance = 0
    for i in range(len(seq1)):
        if seq1[i] != seq2[i]: hammingdistance+=1
    return hammingdistance
    
    
print(checkseqhamming(seq1, seq2))
```

```python
def countpattern(seq, pattern):
    count = 0
    indices = []
    if (len(pattern) > len(seq)): return count
    for i in range(len(seq) - len(pattern) + 1):
        match = True
        for j in range(len(pattern)):
            if seq[i+j] != pattern[j]: match = False
        if (match == True): count+=1; indices.append(i)
    return count, indices


print(countpattern("ACAATACGAAC", "AC"))
```

```python
# levenatneshetien attempt number two
def levevenfeshtein(seq1, seq2):

    grid = [[0 for col in range(len(seq2))] for row in range(len(seq1))]

    for i in range(len(seq1)):
        for j in range(len(seq2)):

            if (seq1[i] == seq2[j]):
                if (i > 0 and j > 0): grid[i][j] = grid[i-1][j-1]
                elif (i > 0): grid[i][j] = grid[i-1][j]
                elif (j > 0): grid[i][j] = grid[i][j-1]

            else:
                if (i > 0 and j > 0): grid[i][j] = min(grid[i-1][j],
                                                       grid[i][j-1],
                                                       grid[i-1][j-1]) + 1
                elif (i > 0): grid[i][j] = grid[i-1][j] + 1
                elif (j > 0): grid[i][j] = grid[i][j-1] + 1
                else: grid[i][j] = 1

    return grid


ans = levevenfeshtein("kelm", "hello")
for row in ans: print(row)
```

```python
def count_kmers(seq, size):
    words = []
    for i in range(len(seq) - size + 1):

        currword = seq[i:i+3]
        index = 0

        for word in words:
            if currword != word[0]: index+=1
            else: break

        if index == len(words): words.append([currword, 1])
        else: words[index][1]+=1

    return words
    

print(count_kmers("GATGAT", 3))
```

```python
# Here, I amplify my count_kmers function from before by adding two things:
# 1. Using a dictionary as that is much more suitable for this problem
# 2. Adding the indices of each kmer alongside its frequency
def count_kmers_dict_with_indices(seq, size):
    
    kmer_dict = {seq[0:size]: [1, [0]]}
    
    for i in range(1, len(seq) - size + 1):
    
        kmer = seq[i:i+size]
    
        if kmer not in kmer_dict: kmer_dict[kmer] = [1, [i]]
        else: kmer_dict[kmer][0]+=1; kmer_dict[kmer][1].append(i)

    return kmer_dict
    

for kmer in count_kmers_dict_with_indices("GATGAT", 3).items(): print(kmer)
```

```python
def complement(seq):
    seq = seq.upper()
    isdna = 'T' in seq
    complementseq = ''
    for n in seq:
        if n == 'A': complementseq += ('T' if isdna else 'U')
        elif n == ('T' if isdna else 'U'): complementseq += 'A'
        elif n == 'C': complementseq += 'G'
        elif n == 'G': complementseq += 'C'
    return complementseq

def reverse(seq):
    if len(seq) > 0: return seq[-1] + reverse(seq[:-1])
    else: return seq

print(reverse(complement("TACAAGCTCATTAGG")))
print(reverse(complement("UCGAUAGCUAGCUAC")))
```

Longest Common Subsequence

```python
def __lcs(seq1, seq2):

    # Generate grid of zeroes with dimensions of (len(seq1) + 1)×(len(seq2) + 1)
    grid = [[0 for col in range(len(seq2) + 1)] for row in range(len(seq1) + 1)]

    # Loop over entire grid with indices i, j
    # Starting from the 2nd row and column
    for i in range(1, len(seq1) + 1):
        for j in range(1, len(seq2) + 1):
            
            # If the immediate top-left letters match...
            if (seq1[i-1] == seq2[j-1]):

                # Add one to corresponding cell grid
                # This resembles incrementing the LCS with this new match
                grid[i][j] = grid[i-1][j-1] + 1

            # Otherwise; if they don't match...
            else:

                # Get the max between left and top
                # This resembles proceeding with the current LCS so far
                grid[i][j] = max(grid[i-1][j], grid[i][j-1])

    # The gist of the code portion that follows is:
    # In order to trace back the subsequence, need to get the points at which...
    # ...the longest sequence was updated. These points are those in the grid...
    # ...surrounded horizontally, vertically, and diagonally by a smaller value.
    
    i = len(seq1); j = len(seq2); lcs = []
    while (i > 0 and j > 0):
        
        # Decrement horizontally as long as grid cells match
        while (j > 0 and grid[i][j] == grid[i][j-1]): j-=1

        # Decrement vertically as long as grid cells match
        while (i > 0 and grid[i][j] == grid[i-1][j]): i-=1

        # Insert once you've reached the corner point
        lcs.insert(0, seq1[i-1])
        
        # Decrement diagonally to go to previous position in subsequence
        i-=1; j-=1

    return ''.join(lcs), grid


def lcs(seq1, seq2, debugging=False):
    lcs, grid = __lcs(seq1, seq2)
    if not debugging: return lcs
    else:
        out =(f'\n\n\nLCS: {lcs}, of length: {len(lcs)}\n')
        out+=("\nGrid:\n")
        grid[0].insert(0, ' ')
        for i, row in enumerate(grid[1:]): row.insert(0, seq1[i])
        grid.insert(0, [' ', ' '] + [i for i in seq2])
        for row in grid:
            for item in row:
                out+=(f' {item} ')
            out+=('\n')
        out+=("\n\n\n")
        return out


print(lcs("GXTXAYB", "AGGTAB", debugging=True)) # GTAB
print(lcs("AGGTAB", "GXTXAYB", debugging=True)) # GTAB
print(lcs("ABCDEF", "CXABEF")) # ABEF
print(lcs("CXABEF", "ABCDEF")) # ABEF
print(lcs("AXCXTXG", "ACGG")) # ACG
print(lcs("ACGG", "AXCXTXG")) # ACG
print(lcs("ABXXEF", "ABFCD")) # ABF
print(lcs("ABFCD", "ABXXEF")) # ABF
print(lcs("ABCD", "ABCD", debugging=True)) # ABCD
print(lcs("AXXBCD", "ABCD")) # ABCD
print(lcs("ABCD", "AXXBCD")) # ABCD
print(lcs("ABYYYCD", "AXXBCDYX")) # ABCD
print(lcs("AXXBCDYX", "ABYYYCD")) # ABCD


# Problem in some test cases such as #5
# This should output ACG, but it outputs CG... problem in tracing back
# ------------------------------------------------------------------------------
# Supposedly fixed after re-writing the tracing back algorithm from scratch
# Not sure what the bug was tbh, but new algorithm is simpler anyway so win-win
```

Pairwise Sequence Alignment via Window Scoring their Dot Plot Matrix

```python
def psaws(seq1, seq2, window_size, window_step_size, acceptance_threshold):

    # Generate initial dot plot placing a dot (•) whenever letters match
    dot_plot = [[' • ' if i == j else '   ' for i in seq1] for j in seq2]

    # Loop over data window-by-window vertically and horizontally
    for i in range(0, len(seq1) - window_size + 1, window_step_size):
        for j in range(0, len(seq2) - window_size + 1, window_step_size):
            
            # Count (diagonally) number of dots in current (i, j) window
            window_dot_count = 0
            for k in range(window_size):
                window_dot_count += '•' in dot_plot[i+k][j+k]
            
            # If number greater than threshold, mark the window's center
            # Note that the marking should not influence other iterations
            # That is why it doesn't add or remove any (non-)existing dots
            # It only surrounds whatever already existed there with '[', ']'
            if window_dot_count >= acceptance_threshold:
                dot_plot[int(i+window_size/2)][int(j+window_size/2)] = \
                f'[{dot_plot[int(i+window_size/2)][int(j+window_size/2)][1]}]'

    # Prepare dot plot for nicer visualization with the sequences then return it
    for i, row in enumerate(dot_plot): row.insert(0, f" {seq2[i]} ")
    dot_plot.insert(0, ['   '] + [f" {i} " for i in seq1])
    return dot_plot


def dot_plot_print(dot_plot):
    for row in dot_plot:
        for item in row:
            print(item, end='')
        print()
    print()


dot_plot_print(psaws("ACCTTGTCCTCTTTGCCC", "ACGTTGACCTGTAACCTC", 9, 3, 4))

dot_plot_print(psaws("ABCDEF", "ABXDXF", 3, 1, 2)) # This is a corner case
# It demonstrates why I don't add a dot upon marking to avoid wrong marks later
```

Global Sequence Alignment via:
1. Maximum Score Matrix
2. Needleman-Wunsch Algorithm

```python
def msm(seq1, seq2):
    
    # Create all-zeroes grid initially with dimensions of the two sequences
    grid = [[0 for col in seq2] for row in seq1]
    
    # Fill first row and column with cumulative match scores only
    for i in range(len(seq1)): grid[i][0] = int(seq1[i] == seq2[0])
    for j in range(len(seq2)): grid[0][j] = int(seq1[0] == seq2[j])

    # Fill the rest of the grid with cumulative match scores + surrounding max
    for i in range(1, len(seq1)):
        for j in range(1, len(seq2)):
            grid[i][j] = int(seq1[i] == seq2[j]) + max(grid[i-1][j],
                                                       grid[i][j-1],
                                                       grid[i-1][j-1])

    return grid


for row in msm("CACTGT", "CATGTT"): print(row)
```

```python
def needleman_wunsch(seq1, seq2, match_score, mismatch_score, gap_score):

    # Create all-zeroes grid initially with dimensions of the two sequences + 1
    grid = [[0 for col in range(len(seq2) + 1)] for row in range(len(seq1) + 1)]
    
    # Fill first row and column with cumulative gap scores
    for i in range(1, len(seq1) + 1): grid[i][0] = grid[i-1][0] + gap_score
    for j in range(1, len(seq2) + 1): grid[0][j] = grid[0][j-1] + gap_score

    # Fill the rest of the grid with cumulative match/mismatch/gap score combos
    for i in range(1, len(seq1) + 1):
        for j in range(1, len(seq2) + 1):
            grid[i][j] = max(grid[i-1][j] + gap_score,
                             grid[i][j-1] + gap_score,
                             grid[i-1][j-1] + (match_score if seq1[i-1] == seq2[j-1] else mismatch_score))


    return grid


for row in needleman_wunsch("CATGT", "ACGCTG", 2, -1, -1):
    for cell in row: print(cell, end='\t')
    print('\n\n')
```
