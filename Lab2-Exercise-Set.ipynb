```python
# Detects if sequence given was RNA or DNA on its own
# Complements it accordingly
# Leaves 'N' instances as-is
# Case-insensitive
def complement(seq):
    compseq = ''
    seq = seq.upper()
    isrna = 'U' in seq
    for nuc in seq:
        if (nuc == 'A'): compseq += ('U' if isrna else 'T')
        elif (nuc == ('U' if isrna else 'T')): compseq += 'A'
        elif (nuc == 'C'): compseq += 'G'
        elif (nuc == 'G'): compseq += 'C'
        else: compseq += 'N'

    return compseq


print(complement('ACGTn'))
print(complement('ACGUn'))
print(complement("ACTGAT"))
```

```python
def reversecomplement(seq):
    return complement(seq)[::-1]


print(reversecomplement("ACTGAT"))
```

```python
def transcription(dnaseq):
    transseq = ''
    dnaseq = dnaseq.upper()
    for nuc in dnaseq:
        if (nuc == 'A'): transseq += 'U'
        elif (nuc == 'T'): transseq += 'A'
        elif (nuc == 'C'): transseq += 'G'
        elif (nuc == 'G'): transseq += 'C'
        else: transseq += 'N'

    return transseq


print(transcription("ACTGAT"))
```

```python
def reversetranscription(dnaseq):
    return transcription(dnaseq)[::-1]


print(reversetranscription("ACTGAT"))
```

```python
# Assume that the input to this function is the coding DNA ***or*** the mRNA
def translate(seq):
    seq = seq.upper().replace('T', 'U') # if given coding DNA, convert into mRNA
    mrna_protein_map = {

        "UUU":"F",  "CUU":"L",  "AUU":"I",  "GUU":"V",
        "UUC":"F",  "CUC":"L",  "AUC":"I",  "GUC":"V",
        "UUA":"L",  "CUA":"L",  "AUA":"I",  "GUA":"V",
        "UUG":"L",  "CUG":"L",  "AUG":"M",  "GUG":"V",
        
        "UCU":"S",  "CCU":"P",  "ACU":"T",  "GCU":"A",
        "UCC":"S",  "CCC":"P",  "ACC":"T",  "GCC":"A",
        "UCA":"S",  "CCA":"P",  "ACA":"T",  "GCA":"A",
        "UCG":"S",  "CCG":"P",  "ACG":"T",  "GCG":"A",

        "UAU":"Y",  "CAU":"H",  "AAU":"N",  "GAU":"D",
        "UAC":"Y",  "CAC":"H",  "AAC":"N",  "GAC":"D",
        "UAA":"*",  "CAA":"Q",  "AAA":"K",  "GAA":"E",
        "UAG":"*",  "CAG":"Q",  "AAG":"K",  "GAG":"E",

        "UGU":"C",  "CGU":"R",  "AGU":"S",  "GGU":"G",
        "UGC":"C",  "CGC":"R",  "AGC":"S",  "GGC":"G",
        "UGA":"*",  "CGA":"R",  "AGA":"R",  "GGA":"G",
        "UGG":"W",  "CGG":"R",  "AGG":"R",  "GGG":"G" 

    }
    protein = ""
    if len(seq) % 3 == 0:
        for i in range(0, len(seq), 3):
            protein += mrna_protein_map[seq[i:i+3]]

    return protein


print(translate("CTAGCTACCCGGATATATCCTAGCTTTTAG"))
```

```python
def hamming(seq1, seq2):
    if len(seq1) != len(seq2): print("ERROR: Sequences must be of same length")
    dist = 0
    for i in range(len(seq1)):
        if seq1[i].upper() != seq2[i].upper(): dist+=1

    return dist


print(hamming(
    "ATCATCATCTAGAGCAGTACGCTGC",
    "GTCATCATCTAGCGCAGTACTCTGT"))
```

```python
def threeframetranslation(seq):
    readings  = "The three possible readings are:"
    readings += '\n'   + seq[0:len(seq) - (len(seq) - 0)%3]
    readings += '\n '  + seq[1:len(seq) - (len(seq) - 1)%3]
    readings += '\n  ' + seq[2:len(seq) - (len(seq) - 2)%3]
    readings += "\n"

    return readings, [
    translate(seq[0:len(seq) - (len(seq) - 0)%3]),
    translate(seq[1:len(seq) - (len(seq) - 1)%3]),
    translate(seq[2:len(seq) - (len(seq) - 2)%3])]


readings, proteins = threeframetranslation("CTAGCTACCCGGATATATCCTAGCTTTTAG")
print(readings, proteins)
```
